```python
videos = glob.glob('./Videos/*')
OUTPUT_PATH = './frames/bg_canny/'
print(videos)
fgbg = cv.bgsegm.createBackgroundSubtractorMOG()
for i in videos:
    path = i
    vid=cv.VideoCapture(path)
    count=0
    print(i)
    folder_name = path.split('.')[1].split('/')[2]
    folder_path = OUTPUT_PATH + folder_name
    print(folder_path,folder_name)
    if os.path.exists(folder_path):
        continue
    os.makedirs(folder_path)
    success=1
    while success:
        if True: #count%3 == 0:
            success, image=vid.read()
            image = fgbg.apply(image)
            image = cv.Canny(image,20,55,apertureSize=3)
            filename = folder_path+"/"+str(count)+"_"+folder_name+".jpg"
            print(filename)
            cv.imwrite(filename, image)
        count+=1
```
